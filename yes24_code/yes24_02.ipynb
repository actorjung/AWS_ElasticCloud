```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'http://www.yes24.com/24/category/bestseller'
```

```python
html = requests.get(url).text
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
books = soup.select('ol > li')
print(len(books))
```

```python
book_list = []
for book in books:
    title = book.select('p > a')[2].text
    price = book.select('p.price > strong')[0].text
    writer_company = book.select('p.aupu')[0].text
    writer = writer_company.split('|')[0]
    company = writer_company.split('|')[1]

    stars = book.select('p > img')[-11:-1]

    star_content = 0
    star_edit = 0
    
    for star in stars[:5]:
        
        if star['src'] == 'http://image.yes24.com/sysimage/interface/smallStarOn.gif':
            star_content += 1

    for star in stars[5:]:
        if star['src'] == 'http://image.yes24.com/sysimage/interface/smallStarOn.gif':
            star_edit += 1
         

    
    # for star in star[5:]:
    #     if star['src'] == 'http://image.yes24.com/sysimage/interface/smallStarOn.gif':
    #         star_edit+=1
    #제목, 가격, 저자, 출판사, 내용 평점, 편집/구성 평점 
    print(title, price,writer, company, star_content,star_edit, sep = '/')
    book_list.append([title, price, writer, company, star_content, star_edit])
```
