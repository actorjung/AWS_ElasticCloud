```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
category_num = {
                "로맨스":"017001046",
                "BL": "017001064",
                "만화":"017001038",
                "판타지 무협": "017001049",
                "소설": "017001045",
                "경제 경영" : "017001047",
                "라이트노벨" : "01700106",
                "에세이 시" : "017001050",
                "인문" : "017001048",
                "사회 정치" : "017001051",
                "자기계발" : "017001054",
                "역사" : "017001052",
                "종교" : "017001053",
                "예술 대중문화" : "017001056",
                "자연과학" : "017001055",
                "가정 살림" : "017001042",
                "건강 취미 여행" : "017001043",
                "어린이 유아" : "017001058",
                "청소년" : "017001059",
                "국어 외국어" : "017001060",
                "IT 모발일" : "017001040",
                "대학교재" : "017001041",
                "수험서 자격증" : "017001061",
                "잡지" : "017001036",
                "해외원서": "017001032",
                "오디오북": "017001032",
                "크레마 (전자책 단말기)": "017001032"
                
}
```

```python
def create_url(category_number):
    url = "http://www.yes24.com/24/category/bestseller?CategoryNumber={0}&sumgb=07&FetchSize=80".format(category_number)

    return url
    
```

```python
def create_url2(category_number):
    url = "http://www.yes24.com/24/category/bestseller?CategoryNumber={0}&sumgb=07&FetchSize=20&PageNumber=2".format(category_number)

    return url
    
```

```python

for key, value in category_num.items():

    crolling_url = create_url(value)
    html = requests.get(crolling_url).text
    soup = BeautifulSoup(html, 'html.parser')
    books = soup.select('td.goodsTxtInfo')
    bookss = []
    
    

    for book in books:
        title = book.select('p:nth-child(1)')[0].text
        price = book.select('p:nth-child(3) > span.priceB')[0].text
        info = book.select('div.aupu')[0].text
        
        writer = info.split('|')[0]

        publisher = info.split('|')[1]

        date = info.split('|')[2]

        stars = book.select('p.review > img')
        
        star_content = 0
        star_edit = 0

        if stars==None:
                    
                star_content = 0
                star_edit = 0
                
        else:

        
            for star in stars[:5]:
                
                if star['src'] == 'http://image.yes24.com/sysimage/interface/StarOn.gif':
                    
                    star_content += 1

            for star in stars[5:]:

                if star['src'] == 'http://image.yes24.com/sysimage/interface/StarOn.gif':
        
                    star_edit += 1
        

        #print(key,title, writer, publisher, date, star_content,star_edit)
        bookss.append([key,title,writer, publisher, date,price, star_content,star_edit]) # 80 add

    crolling_url2 = create_url2(value)
    html2 = requests.get(crolling_url2).text
    soup2 = BeautifulSoup(html2, 'html.parser')
    books2 = soup2.select('td.goodsTxtInfo')

        
    for book in books2:
        title = book.select('p:nth-child(1)')[0].text
        price = book.select('p:nth-child(3) > span.priceB')[0].text
        info = book.select('div.aupu')[0].text
        
        writer = info.split('|')[0]

        publisher = info.split('|')[1]

        date = info.split('|')[2]

        stars = book.select('p.review > img')
        
        star_content = 0
        star_edit = 0

        if stars==None:
                    
                star_content = 0
                star_edit = 0
                
        else:

        
            for star in stars[:5]:
                
                if star['src'] == 'http://image.yes24.com/sysimage/interface/StarOn.gif':
                    
                    star_content += 1

            for star in stars[5:]:

                if star['src'] == 'http://image.yes24.com/sysimage/interface/StarOn.gif':
        
                    star_edit += 1
        

        #print(key,title, writer, publisher, date, star_content,star_edit)
        bookss.append([key,title,writer, publisher, date,price, star_content,star_edit]) # 20 add


        

    df = pd.DataFrame(bookss, columns=['genre','title','writer', 'company','date','price' ,'star_content','star_edit'])

    df['date'] = df['date'].str.replace(pat=r'[^A-Za-z0-9가-힣]', repl=r' ', regex=True)

    df['writer']=df['writer'].str.replace(pat=r'[^A-Za-z0-9가-힣]', repl=r' ', regex=True)
    df['title'] = df['title'].str.replace(pat=r'[^A-Za-z0-9가-힣]', repl=r' ', regex=True)
    df['company'] = df['company'].str.replace(pat=r'[^A-Za-z0-9가-힣]', repl=r' ', regex=True)
    df['rank'] = df.index + 1
    df = df[['rank','genre','title','writer', 'company','date','price' ,'star_content','star_edit']]
    df.to_csv('csv_file\yes24_{0}.csv'.format(key), index= False, encoding= 'utf-8')


        

    
    
    

    
    


    

```
