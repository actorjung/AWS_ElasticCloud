```python
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://libebook.smu.ac.kr/FxLibrary/product/list/?category_type=book&cateopt=best' 
res = requests.get(url)
print(res)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
book_list = []
books = soup.select('li.item')
books = books[0:20]
for book in books:
  title = book.select_one('div.subject > a').text  #책 제목
  author = book.select_one('div.info >  ul.i1 > li').text  #저자
  date = book.select('div.info > ul.i1 > li')[2].text  #등록일
  ctg = book.select_one('div.category > span').text #장르
  cum = book.select('div.info > ul.i2> li')[2].text  #대출횟수
 # rec = book.select('div.info > ul.i2 > li')[3].text  #추천수
  print(title,author,date,ctg)
  book_list.append([title,author,date,ctg])
```

```python
cat_list = []
categorys = soup.select('li.item > div.category')
for category in categorys:
  ctg = category.select_one('span').text #장르
  cat_list.append(ctg)
```

```python
num_list = []
nums = soup.select('li.item')
for num in nums:
  cum = num.select_one('div.info > ul.i2> li')[2].text  #대출횟수
  rec = num.select_one(' div.info > ul.i2 > li')[3].text  #추천수
  print(cum,rec)
  num_list.append([cum,rec])
```

```python
num_list = []
nums = soup.select('li.item > div.info > ul.i2')
for num in nums:
  cum = num.select('li')[2].text   #대출횟수
  rec = num.select('li')[3].text   #추천수
  num_list.append([cum,rec])
```
